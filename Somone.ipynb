```python
import nibabel as nib
```

```python
epi_img = nib.load('/Users/hxia/Downloads/someones_epi.nii.gz')
```

```python
epi_img_data = epi_img.get_data()
epi_img_data.shape
```

```python
import matplotlib.pyplot as plt
```

```python
def show_slices(slices):
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap ='gray',origin='lower')
```

```python
slice_0 = epi_img_data[26,:,:]
slice_1 = epi_img_data[:,30,:]
slice_2 = epi_img_data[:,:,16]
```

```python
show_slices([slice_0,slice_1,slice_2])
plt.suptitle('Center slices for EPI image')
```

```python
anat_img = nib.load('/Users/hxia/Downloads/someones_anatomy.nii.gz')
anat_img_data = anat_img.get_data()
anat_img_data.shape
show_slices([anat_img_data[28, :, :],anat_img_data[:, 33, :],anat_img_data[:, :, 28]])
plt.suptitle("Center slices for anatomical image")
```
